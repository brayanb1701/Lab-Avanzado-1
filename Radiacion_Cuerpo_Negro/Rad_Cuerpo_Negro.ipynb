```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Cargar el archivo Excel
file_path = 'cubo de leslie.xlsx'
data = pd.read_excel(file_path)

# Convertir temperaturas de Celsius a Kelvin
data['T(K)'] = data['T'] + 273.15

# Constante T_0 en Kelvin
T0_K = 24 + 273.15

# Cálculo de T^4 - T0^4
data['T^4 - T0^4'] = data['T(K)']**4 - T0_K**4

# Conversión de voltajes
voltaje_columns = ['Cara negra', 'Cara blanca', 'Cara espejo', 'Cara metálica gris']
data[voltaje_columns] = data[voltaje_columns] * 10**-4

# Preparar la gráfica
plt.figure(figsize=(10, 10))

# Ajuste lineal y gráfica para cada columna de voltaje
for col in voltaje_columns:
    # Realizar ajuste lineal
    slope, intercept, r_value, p_value, std_err = linregress(data['T^4 - T0^4'], data[col])

    # Generar valores de y para la línea ajustada
    line = slope * data['T^4 - T0^4'] + intercept

    # Graficar puntos y línea ajustada
    plt.scatter(data['T^4 - T0^4'], data[col], label=f'{col}')
    plt.plot(data['T^4 - T0^4'], line, label=f'Ajuste {col} (R²={r_value**2:.2f})')

# Etiquetas y leyenda
plt.xlabel('$T^4 - T_0^4$ [K^4]')
plt.ylabel('Voltaje [V]')
plt.title('Cubo de Leslie (Voltaje vs. $T^4 - T_0^4$)')
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Cargar el archivo Excel
file_path = 'cubo de leslie.xlsx'
data = pd.read_excel(file_path)

# Convertir temperaturas de Celsius a Kelvin
data['T(K)'] = data['T'] + 273.15

# Constante T_0 en Kelvin
T0_K = 24 + 273.15

# Cálculo de T^4 - T0^4
data['T^4 - T0^4'] = data['T(K)']**4 - T0_K**4

# Conversión de voltajes
voltaje_columns = ['V1', 'V2', 'V3', 'V4']
names = ['Cara negra', 'Cara blanca', 'Cara espejo', 'Cara metálica gris']
data[voltaje_columns] = data[voltaje_columns] * 10**-4

# Colores personalizados para cada curva
colors = ['purple', 'gray', 'black', 'red']

# Preparar la gráfica
plt.figure(figsize=(10, 10))

# Ajuste lineal y gráfica para cada columna de voltaje
for name, col, color in zip(names, voltaje_columns, colors):
    # Realizar ajuste lineal
    slope, intercept, r_value, p_value, std_err = linregress(data['T^4 - T0^4'], data[col])

    # Generar valores de y para la línea ajustada
    line = slope * data['T^4 - T0^4'] + intercept

    # Graficar puntos y línea ajustada con color personalizado
    plt.scatter(data['T^4 - T0^4'], data[col], label=f'{name}')
    plt.plot(data['T^4 - T0^4'], line, label=f'Fit ({slope:.5e}x+{intercept:.5f})')

# Etiquetas y leyenda
plt.xlabel('$T^4 - T_0^4$ ($K^4$)', fontsize=14)
plt.ylabel('Voltaje (V)', fontsize=14)
plt.title('Cubo de Leslie (Voltaje vs. $T^4 - T_0^4$)', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Load the provided Excel file
file_path = 'datos rad cuerpo negro.xlsx'
data = pd.read_excel(file_path)

# Convert temperatures from Celsius to Kelvin
data['T_asc [K]'] = data['T_asc [C]'] + 273.15
data['T_desc [K]'] = data['T_desc [C]'] + 273.15

# Convert voltage to Volts (multiply by 10^-4)
data['V_asc [V]'] = data['V_asc [V e-4]'] * 1e-4
data['V_desc [V]'] = data['V_desc[V e-4]'] * 1e-4

# Constant T0 in Kelvin (24°C)
T0_K = 24 + 273.15

# Calculate T^4 - T0^4 for both ascending and descending temperatures
data['T_asc^4 - T0^4'] = data['T_asc [K]'] ** 4 - T0_K ** 4
data['T_desc^4 - T0^4'] = data['T_desc [K]'] ** 4 - T0_K ** 4

# Remove NaN values from descending data
data_desc_cleaned = data.dropna(subset=['T_desc [C]', 'V_desc [V]'])

# Perform linear regression for ascending data
slope_asc, intercept_asc, r_value_asc, _, _ = linregress(data['T_asc^4 - T0^4'], data['V_asc [V]'])

# Calculate the fitted line values for ascending data
line_asc = slope_asc * data['T_asc^4 - T0^4'] + intercept_asc

# Perform linear regression for cleaned descending data
slope_desc_cleaned, intercept_desc_cleaned, r_value_desc_cleaned, _, _ = linregress(
    data_desc_cleaned['T_desc^4 - T0^4'], data_desc_cleaned['V_desc [V]'])

# Calculate the fitted line values for cleaned descending data
line_desc_cleaned = slope_desc_cleaned * data_desc_cleaned['T_desc^4 - T0^4'] + intercept_desc_cleaned

# Plotting the data with linear fits
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# Ascending data with fit
plt.scatter(data['T_asc^4 - T0^4'], data['V_asc [V]'], color='blue', label='Ascending')
plt.plot(data['T_asc^4 - T0^4'], line_asc, color='blue', linestyle='dashed',
         label=f'Ascending Fit (slope: {slope_asc:.5e}, $R^2$: {r_value_asc**2:.3f})')

# Cleaned descending data with fit
plt.scatter(data_desc_cleaned['T_desc^4 - T0^4'], data_desc_cleaned['V_desc [V]'], color='red', label='Descending (Cleaned)')
plt.plot(data_desc_cleaned['T_desc^4 - T0^4'], line_desc_cleaned, color='red', linestyle='dashed',
         label=f'Descending Fit (Cleaned) (slope: {slope_desc_cleaned:.5e}, $R^2$: {r_value_desc_cleaned**2:.3f})')

# Labels and title
plt.xlabel('$T^4 - T_0^4$ ($K^4$)', fontsize=14)
plt.ylabel('Voltaje (V)', fontsize=14)
plt.title('Voltaje vs. $T^4 - T_0^4$ (Ascenso y Descenso) con Ajuste Lineal', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()

# Show the plot
plt.show()
```
