```python
I = np.array([0.86, 1.09, 1.35, 1.63, 1.95,2.23])
V = np.array([103, 125.7, 132.8, 151.6, 173.3, 186.1])

# Calcula I al cuadrado
I_squared = I**2

# Realiza un ajuste lineal
slope, intercept = np.polyfit(I_squared, V, 1)

# Crea un rango para la línea de ajuste
I_fit = np.linspace(I_squared.min(), I_squared.max(), 100)
V_fit = slope*I_fit + intercept

# Grafica los datos y la línea de ajuste
plt.scatter(I_squared, V, label="Datos")
plt.plot(I_fit, V_fit, 'r', label=f"Ajuste lineal: y={slope:.2f}x + {intercept:.2f}")
plt.title("V en función de I^2 para R1 = 2 cm")
plt.xlabel("I^2 (A^2)")
plt.ylabel("V (V)")
plt.legend()
plt.grid(True)
plt.show()

e=slope*2*0.2**2/(0.716*u0*154*2e-2)**2
error=abs(e-e_real)/e_real*100
print(e)
print(error)
print(f"Pendiente de la curva: {slope:.2f}")
```
